```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import sklearn as sk
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [
    'To be or not to be? that is the question.',
    'But, to know a man well, were to know himself.',
    'Knowing yourself is the beginning of all wisdom.',
    'To know yourself is to know others.'
    ]
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(corpus)
```

```python
vectorizer.get_feature_names_out()
```

```python
print(X.toarray())
```

```python
pd.DataFrame(X.A, columns=vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer2 = TfidfVectorizer()
```

```python
X1 = vectorizer2.fit_transform(corpus) 
```

```python
vectorizer2.get_feature_names_out()
```

```python
print(X1.toarray())
```

```python
pd.DataFrame(np.round(X1.A,2), columns=vectorizer2.get_feature_names_out())
```

```python
vectorizer3 = CountVectorizer(analyzer='word', ngram_range=(2, 2))
```

```python
X2 = vectorizer3.fit_transform(corpus)
```

```python
vectorizer3.get_feature_names_out()
```

```python
print(X2.toarray())
```

```python
pd.DataFrame(X2.A, columns=vectorizer3.get_feature_names_out())
```
